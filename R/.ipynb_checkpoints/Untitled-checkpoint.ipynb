```r
library(caret)
library(klaR) # Used for kmodes (an generalized version of the kmeans supporting nominal atributes)
library(RWeka) # Used for cobweb
library(mclust)
library(kohonen)
library(RColorBrewer)

library("arules"); 
library("arulesViz");
```

```r
# https://www.rdocumentation.org/packages/klaR/versions/0.6-14/topics/GermanCredit
credit <- read.csv('german-credit-balanced.csv', sep=',')
head(credit)

credit.disc <- read.csv('german-credit-discretized.csv', sep=',')
head(credit.disc)

# dummify the data
dmy <- dummyVars(" ~ .", data = credit)
credit.dmy <- data.frame(predict(dmy, newdata = credit))
head(credit.dmy)

# testing data
credit.test <- read.csv('newUsers.csv')
credit.test
```

# K-Means
## Definiendo el número de clusters a detectar

Determinar el parámetro K es uno de los problemas de las técnicas no supervisadas, existen diversos métodos para hallar un valor optimo, como por ejemplo maximizar el Bayesian Information Criterion, sin embargo para nuestro caso se utilizó un valor arbitrario de **k=2** debido a que en nuestro caso tenemos dos clases


```r
# https://www.r-bloggers.com/k-means-clustering-in-r/
# Kmode is like a kmeans for cluster numerical and categorical data https://datascience.stackexchange.com/questions/22/k-means-clustering-for-mixed-numeric-and-categorical-data

# trains, second argument is k=2
clusterMdl <- kmodes(credit, 2)
# plots
plot(jitter(credit.dmy$class.good), credit.cluster$cluster)
# predict new users
#credit.cluster


# Bind credit with cluster https://discuss.analyticsvidhya.com/t/how-to-add-a-column-to-a-data-frame-in-r/3278/2
credit.clusters <- cbind(credit, clusterMdl$cluster)
colnames(credit.clusters)[22] <- 'cluster'
credit.clusters[,22] <- as.factor(credit.clusters$cluster)
sapply(credit.clusters, class)
head(credit.clusters)

# Learns the cluster of each instance for predict new instances
trctrl <- trainControl(method = "repeatedcv", number = 5, repeats = 3)
svmlinear <- train(cluster ~ .,
                 data = credit.clusters,
                 method = "svmLinear",
                 trControl=trctrl,
                 tuneLength = 10
                 )
svmlinear
```

## Prediciendo el cluster al que pertenecen 5 usuarios nuevos

```r
credit.newclients <- predict(svmlinear, newdata=credit.test)
credit.predicted <- cbind(credit.test, credit.newclients)
colnames(credit.predicted)[22] <- 'class'
credit.predicted
```

# Self-Organising Maps

```r
# https://www.r-bloggers.com/self-organising-maps-for-customer-segmentation-using-r/
# Self-Organising Maps (SOMs) are an unsupervised data visualisation technique that can be used to visualise high-dimensional data sets in lower (typically 2) dimensional representations.

# Create a training data set (rows are samples, columns are variables
# Here I am selecting a subset of my variables available in "data"
data_train <- credit[, c(2,5,8,11)]

# Change the data frame with training data to a matrix
# Also center and scale all variables to give them equal importance during
# the SOM training process. 
data_train_matrix <- as.matrix(scale(data_train))

# Create the SOM Grid - you generally have to specify the size of the 
# training grid prior to training the SOM. Hexagonal and Circular 
# topologies are possible
som_grid <- somgrid(xdim = 20, ydim=20, topo="hexagonal")

# Finally, train the SOM, options for the number of iterations,
# the learning rates, and the neighbourhood are available
som_model <- som(data_train_matrix,
        grid=som_grid,
        rlen=100,
        alpha=c(0.05,0.01),
        keep.data = TRUE)
```

```r
plot(som_model, type="changes")
```

```r
# This is required for access the codes inside som plot
codes <- as.matrix(som_model$codes[[1]])
plot(som_model, type = "property", property = codes[,1], main=colnames(data_train)[1])
plot(som_model, type = "property", property = codes[,2], main=colnames(data_train)[2])
plot(som_model, type = "property", property = codes[,3], main=colnames(data_train)[3])
plot(som_model, type = "property", property = codes[,4], main=colnames(data_train)[4])
```

```r
mydata <- codes 
wss <- (nrow(mydata)-1)*sum(apply(mydata,2,var)) 
for (i in 2:15) {
  wss[i] <- sum(kmeans(mydata, centers=i)$withinss)
}
# plot(wss)
## use hierarchical clustering to cluster the codebook vectors
som_cluster <- cutree(hclust(dist(codes)), 6)
# plot these results:
plot(som_model, type="mapping", bgcol = '#a29bfe', main = "Clusters") 
add.cluster.boundaries(som_model, som_cluster)
```

# Reglas de asociación, Apriori

```r
# https://www.r-bloggers.com/association-rule-learning-and-the-apriori-algorithm/
credits.apriori = as(credit.disc, "transactions"); # This method requires the atributes are discretized
rules = apriori(credits.apriori, parameter=list(support=0.01, confidence=0.5));
rules
```

```r
subrules = rules[quality(rules)$confidence > 0.98];
subrules
plot(subrules, measure=c("support","lift"), shading="confidence", engine='interactive');
```
